```python
def AQI_range(x):
    if(x<=50):
        return "Good"
    elif(x>50 and x<100):
        return "Moderate"
    elif(x>100 and x<200):
        return "Poor"
    elif(x>200 and x<300):
        return "unhealthy"
    elif(x>300 and x<400):
        return "severe"
    else:
        return "hazardous"
```
